Perfume E-Commerce Market Analysis



```python
#Import important libraries

import pandas as pd
```

```python
# Load the dataset for men's and women's perfumes
mens_perfume_df = pd.read_csv('ebay_mens_perfume.csv')
womens_perfume_df = pd.read_csv('ebay_womens_perfume.csv')
```

```python
#Look at data 
mens_perfume_df.head(5)
```

```python
#Look at data 
womens_perfume_df.head(5)
```

```python
#Count occurances of different brands in women parfumes 
counts = {}

for value in womens_perfume_df['brand']:
    counts[value] = counts.get(value, 0) + 1

print(counts)
print(counts['Versace'])
```

```python
#Count occurances of different brands in women parfumes 
counts = {}

for value in mens_perfume_df['brand']:
    counts[value] = counts.get(value, 0) + 1

print(counts)
print(counts['Versace'])
```
